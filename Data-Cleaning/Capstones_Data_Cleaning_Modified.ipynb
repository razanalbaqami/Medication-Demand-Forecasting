 ## We are not finsifed yet, the DSP_Qty	has 0 values as nan we need to calculate that 

```python
def calculate_DSP_MissingValues(DSP_inputs): #takes a dataframe and return a column
    
    DSP_output = [] 
    # Fisrt step change the duration to days values
    conditions = [ 
    (DSP_inputs['DURATION_TYPE'] == 'Week'),
    (DSP_inputs['DURATION_TYPE'] == 'Hour')]
    
    values = [
        DSP_inputs['DURATION_VALUE']*7, 
        DSP_inputs['DURATION_VALUE']/24]
    
    DSP_inputs['DURATION_VALUE'] = np.select(conditions, values, default=DSP_inputs['DURATION_VALUE'])
    #del DSP_inputs['DURATION_TYPE']
    
    # Second step change the frequency to frequency in days 
    conditions = [ 
    (DSP_inputs['FREQUENCY_TYPE'] == 'Week'),
    (DSP_inputs['FREQUENCY_TYPE'] == 'Hour')]
    values = [
        DSP_inputs['FREQUENCY_VALUE']*(1/7), 
        24/DSP_inputs['FREQUENCY_VALUE']]
    
    DSP_inputs['FREQUENCY_VALUE'] = np.select(conditions, values, default=DSP_inputs['FREQUENCY_VALUE'])
    #del DSP_inputs['FREQUENCY_TYPE']
    
    # Third step is to calculte the dsp using the dose and concentration
    # in the original dataframe of specific drug that is of type tablet 
    
    DSP_inputs['DOSE'] = FUROSEMIDE_df['DOSE'].astype('float64')
    DSP_inputs['CONCENTRSATION'] = FUROSEMIDE_df['CONCENTRSATION'].astype('float64')
    DSP_output =((DSP_inputs['DOSE']  / DSP_inputs['CONCENTRSATION']) * DSP_inputs['DURATION_VALUE'] ) * DSP_inputs['FREQUENCY_VALUE']
    
    return DSP_output
    
```

```python
# 1st setp create a column that takes Frequency feature and generate a new coulmn for the type of frequency 
conditions = [ 
    ((FUROSEMIDE_df['FREQUENCY'].str.find('daily') != -1) | (FUROSEMIDE_df['FREQUENCY'].str.find('Daily') != -1) | (FUROSEMIDE_df['FREQUENCY'].str.find('d') != -1)),
    ((FUROSEMIDE_df['FREQUENCY'].str.find('week') != -1) | (FUROSEMIDE_df['FREQUENCY'].str.find('Weekly') != -1)),
     ((FUROSEMIDE_df['FREQUENCY'].str.find('hours') != -1) | (FUROSEMIDE_df['FREQUENCY'].str.find('hour') != -1) | (FUROSEMIDE_df['FREQUENCY'].str.find('H') != -1)),
]

values = [
    'Day',
    'Week',
    'Hour'
]

FREQUENCY_TYPE = np.select(conditions, values, default='-')
```

```python
# split the frequency to extract the numeric and categorical values
fr_features=FUROSEMIDE_df['FREQUENCY'].str.split(' ', expand=True).iloc[:,:2]
fr_features
```

```python
#change the categorical frequency into numric values
conditions = [ 
    (fr_features[0].isin(['daily','at'])),
    (fr_features[0].isin(['twice','Twice'])),
    (fr_features[0].isin(['three','Three'])),
    (fr_features[0].isin(['four'])),
    (fr_features[1].isin(['hour']))
]

values = [
    1, 
    2,
    3,
    4,
    24
]

fr_features[1] = fr_features[1].str.extract('(\d+)')
```

```python
d = np.select(conditions, values)
d = dict(enumerate(d.flatten(), 0))
```

```python
fr_features[1].fillna(d, inplace=True)
```

```python
du_features=FUROSEMIDE_df['DURATION'].str.split(' ', expand=True).iloc[:,1:3]
du_features[2] = du_features[2].str.replace(r'\([^)]*\)', '')

input_df = du_features.rename({1: 'DURATION_VALUE', 2: 'DURATION_TYPE'}, axis=1)
```
